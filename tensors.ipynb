### Matrix operations

```python
torch.manual_seed(42)

f = torch.randint(size=(3,3), low=10, high=99, dtype=torch.float32)
g = torch.randint(size=(3,3), low=10, high=99, dtype=torch.float32)

print(f)
print(g)
```

```python
# matrix multiplication
torch.matmul(f,g)
```

```python
# dot product
v1 = torch.tensor([1,2])
v2 = torch.tensor([3,4])
torch.dot(v1,v2)
```

```python
# transpose
torch.transpose(g,0,1) # row to column 
```

```python
# determinant
torch.det(g)
```

```python
torch.inverse(g)
```
